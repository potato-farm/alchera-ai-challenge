```python
import cv2
import matplotlib.pyplot as plt
import os

TARGETDIR = "../input/val_mask" # valid mask 위치
```

```python
ls = ["TTA","noTTA"] #결과물들 폴더이름 쓰기
ls.append(TARGETDIR)
imgList = os.listdir(ls[0])
```

```python
# imgId = "001_01.png"
imgId = imgList[100] #번호 or 위에 파일이름

imgs = []
for dirPath in ls:
	filePath = dirPath+"/"+imgId
	if "val" in dirPath: 
		s = imgId.split(".")
		filePath = dirPath+"/"+".".join([s[0],"grayscale",s[1]])

	img = cv2.imread(filePath)
	for i in range(14,0,-1):
		img[img==i] = i*14

	imgs.append(img)

plt.figure(figsize=(100,200))
plt.imshow(cv2.hconcat(imgs))
```
