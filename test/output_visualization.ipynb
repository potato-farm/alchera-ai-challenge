```python
import cv2
import matplotlib.pyplot as plt
import os
import shutil

TARGETDIR = "../input/val_mask_visual" # 시각화용으로 흑백채널 라벨링값을 14씩곱해줄것임
```

```python
if not os.path.isdir(TARGETDIR):
	shutil.copytree("../input/val_mask",TARGETDIR)
	imgList = os.listdir(TARGETDIR)
	for img in imgList:
		path = TARGETDIR+"/"+img
		image = cv2.imread(path)
		for i in range(14,0,-1):
				image[image==i] = i*14
		cv2.imwrite(path,image)
```

```python
ls = ["512","5122"] #현재 dir 안에있는 결과들 폴더이름 쓰기
ls.append(TARGETDIR)
imgList = os.listdir(ls[0])
```

```python
# imgId = "001_01.png"
imgId = imgList[300] #번호로하등가 위에 이름으로하등가

imgs = []
for dirPath in ls:
	filePath = dirPath+"/"+imgId
	if "val" in dirPath: 
		s = imgId.split(".")
		filePath = dirPath+"/"+".".join([s[0],"grayscale",s[1]])

	imgs.append(cv2.imread(filePath))

plt.figure(figsize=(100,200))
plt.imshow(cv2.hconcat(imgs))
```
